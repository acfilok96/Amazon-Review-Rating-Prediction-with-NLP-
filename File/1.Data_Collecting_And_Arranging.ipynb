```python
## Import data reading libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Data collection function for each classes

def data_collection_according_to_class(data_set,class_name,count_val=100):
  corpus_x,corpus_y = [],[]
  text = data_set.Text
  score = data_set.Score
  count = 0
  for i in range(len(score)):
    if(count == count_val):
      break
    if(len(text[i])!=0):
      if(score[i]==class_name):
        corpus_x.append(text[i])
        corpus_y.append(score[i])
        count = count + 1
  return np.array(corpus_x),np.array(corpus_y)
```

```python
## Collected data for each classes
## For each classes , we are collecting 20,000 data

count = 20000

text_1_class,score_1_class = data_collection_according_to_class(amazon_data,1,count)
text_2_class,score_2_class = data_collection_according_to_class(amazon_data,2,count)
text_3_class,score_3_class = data_collection_according_to_class(amazon_data,3,count)
text_4_class,score_4_class = data_collection_according_to_class(amazon_data,4,count)
text_5_class,score_5_class = data_collection_according_to_class(amazon_data,5,count)
```

```python
## Putting all data in a corpus

text = np.concatenate([text_1_class,text_2_class,text_3_class,text_4_class,text_5_class])
score = np.concatenate([score_1_class,score_2_class,score_3_class,score_4_class,score_5_class])
print(text.shape,score.shape)
```

```python
## Make data frame for text and score features

temp = pd.DataFrame({"Text" : text, "Score" : score})
temp.head(5)
```

```python
## Save the data as csv file

temp.to_csv("/content/drive/MyDrive/Amazon/Amazon_100000_data/Amazon_data.csv", index = False)
```
