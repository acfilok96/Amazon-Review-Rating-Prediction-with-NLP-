```python
## Import Libraries For Text Processing

import re, string, random, math

## import Libraries for text processing

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
## Import Libraries For Text Processing

import nltk
nltk.download('all')
```

```python
## Punctuation

print(string.punctuation)
```

```python
## Punctuation

table = str.maketrans('', '', string.punctuation)
# stripped = [w.translate(table) for w in words]
print(table)
```

```python
## Data cleaning function

def data_cleaning(data):
    # remove punctuation from text part 1
    data = re.sub('[^a-zA-Z]+',' ',data)
    # convert into lower letter
    data = data.lower()
    # split text
    data = data.split()
    # remove the words of length one or two
    data = [word for word in data if len(word) > 2]
    # remove punctuation from text part 2
    data = [w.translate(table) for w in data]
    # remove stopword from text
    data = [i for i in data if not i in set(stopwords.words('english'))]
    # lemmatize the text words
    data = ' '.join([WordNetLemmatizer().lemmatize(i) for i in data])
    return data
```

```python
## Clean text 

text = amazon_data.Text
text = text.map(lambda i: data_cleaning(i))
text
```

```python
## score 

score = amazon_data.Score
```

```python
## Make data frame for text and score features

temp = pd.DataFrame({"Text" : text, "Score" : score})
temp.head(5)
```
